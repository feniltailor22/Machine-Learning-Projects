```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import make_blobs
```

```python
data= make_blobs(n_samples=200,n_features=2, centers=4, cluster_std=1.8, random_state=101)
#make_blobs is used to create multiple datapoints close to each other.
```

```python
data
```

```python
plt.scatter(data[0][:,0],data[0][:,1],c=data[1]) #with answers
#data[0][:0]=all the rows in the first column from the first feature array i.e. input datapoints
#data[0][:1]=all the rows in the second column from the first feature array i.e. input datapoints
#data[1]= data from the second feature array(coloring based on cluster centers i.e. answers)
```

```python
plt.scatter(data[0][:,0],data[0][:,1]) #without answers
#data[0][:0]=all the rows in the first column from the first feature array i.e. input datapoints
#data[0][:1]=all the rows in the second column from the first feature array i.e. input datapoints
```

```python
from sklearn.cluster import KMeans
```

```python
km= KMeans(n_clusters=4)
```

```python
km.fit(data[0])
```

```python
km.cluster_centers_
```

```python
km.labels_ 
#predicted labels(output) 
```

```python
data[1] #original labels(output)
```

```python
#Comparing the results by creating subplots
fig, (ax1, ax2)= plt.subplots(nrows=1,ncols=2,sharey=True,figsize=(10,6))
#nrows, ncols= Number of rows/columns of the subplot grid

#Subplot-1 
ax1.set_title('KMeans_Prediction')
ax1.scatter(data[0][:,0],data[0][:,1],c=km.labels_)

#Subplot-2
ax2.set_title('Original_Output')
ax2.scatter(data[0][:,0],data[0][:,1],c=data[1])
```

```python
#Find optimal number of clusters and visualize it
import numpy as np
Ks = 20

mse = np.zeros((Ks-1))

for n in range(1,Ks):

    # set number of clusters

    kclusters = n

    # run k-means clustering

    kmeans = KMeans(n_clusters=kclusters, random_state=0, n_init = 15).fit(data[0])

    mse[n-1] = kmeans.inertia_

print(mse)

plt.plot(range(1,Ks),mse)

plt.xlabel("Optimal number of K")

plt.ylabel("MSE")

plt.title("K selection")

plt.show()
```
